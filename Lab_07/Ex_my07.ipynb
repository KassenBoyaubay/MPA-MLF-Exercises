```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.optimizers import SGD, Adam
from keras.datasets import cifar10
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
from keras import regularizers
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import cv2

font = {'weight' : 'bold',
        'size'   : 12}

matplotlib.rc('font', **font)
```

```python
def display_random_images(Npoints: int = 1000, Period: int = 6, Time: int = 6, Amp: int = 1) -> list[np.array, np.array]:
  step = Time/Npoints
  x = np.arange(0,Time,step)
  y = Amp*np.sin(x*2*np.pi/Period)
  return [x, y]
```

```python
Npoints = 1000
Period = 3
Time = 12
Amp = 1
[x, y] = display_random_images(Npoints, Period, Time, Amp)
plt.plot(x,y)
```

```python
x = np.expand_dims(x, -1)
y = np.expand_dims(y, -1)
```

```python
model = Sequential()
model.add(Dense(10, input_shape=(1,), activation='tanh'))
model.add(Dense(5, activation='tanh'))
model.add(Dense(1, activation='tanh'))

model.compile(loss='mean_squared_error', optimizer='SGD', metrics=['mean_squared_error'])

history = model.fit(x, y, epochs=10, batch_size=1, verbose=1)
```

```python
[x_test, y_test] = display_random_images(Npoints, Period*2, Time, Amp)
x_test = np.expand_dims(x_test, -1)
y_test = np.expand_dims(y_test, -1)
y_pred = model.predict(x_test)
```

```python
plt.plot(x_test,y_test)
plt.plot(x_test,y_pred)
```
